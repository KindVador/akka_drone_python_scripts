# Flight Log Analyzer

```python
import math
import pyulog
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = (20, 10)
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Tahoma']
rcParams['figure.subplot.top'] = 0.977
rcParams['figure.subplot.bottom'] = 0.073
rcParams['figure.subplot.left'] = 0.036
rcParams['figure.subplot.right'] = 0.99
rcParams['figure.subplot.hspace'] = 0.2
rcParams['figure.subplot.wspace'] = 0.2

plt.style.use('ggplot')

%matplotlib inline

__rad2deg__ = 180.0 / math.pi
```

```python
def quaternion2euler(q0, q1, q2, q3):
    """
    Quaternion rotation from NED earth frame to XYZ body frame

    https://en.wikipedia.org/wiki/Conversion_between_quaternions_and_Euler_angles

    Args:
        q0 (float): first quaternion.
        q1 (float): second quaternion.
        q2 (float): third quaternion.
        q3 (float): fourth quaternion.

    Returns:
        tuple: euler angles in radians (roll, pitch, yaw)
    """
    roll = math.atan2(2*(q0*q1 + q2*q3), 1 - 2*(q1*q1 + q2*q2))
    pitch = math.asin(2*(q0*q2 - q3*q1))
    yaw = math.atan2(2*(q0*q3 + q1*q2), 1 - 2*(q2*q2 + q3*q3))
    return roll, pitch, yaw
```

```python
log_file = pyulog.ULog(r'/Users/florian/Downloads/23_43_25.ulg')
vehicle_attitude = log_file.get_dataset('vehicle_attitude')
actuator_outputs = log_file.get_dataset('actuator_outputs')
vehicle_global_position = log_file.get_dataset('vehicle_global_position')
vehicle_command = log_file.get_dataset('vehicle_command')
```

```python
# vehicle_attitude
vehicle_attitude = log_file.get_dataset('vehicle_attitude')
time_data = vehicle_attitude.data['timestamp']
q0 = vehicle_attitude.data['q[0]']
q1 = vehicle_attitude.data['q[1]']
q2 = vehicle_attitude.data['q[2]']
q3 = vehicle_attitude.data['q[3]']
roll = []
pitch = []
yaw = []
for i, _ in enumerate(time_data):
    r, p, y = quaternion2euler(q0[i], q1[i], q2[i], q3[i])
    roll.append(r * __rad2deg__)
    pitch.append(p * __rad2deg__)
    yaw.append(y * __rad2deg__)
fig, axs = plt.subplots(3, 1, sharex='all')
axs[0].plot(time_data, roll)
axs[0].set_ylabel('Roll')
axs[0].grid(True)
axs[1].plot(time_data, pitch)
axs[1].set_ylabel('Pitch')
axs[1].grid(True)
axs[2].plot(time_data, yaw)
axs[2].set_ylabel('Yaw')
axs[2].grid(True)
fig.tight_layout()
#fig.savefig(f"vehicle_attitude2.pdf", dpi=None, facecolor='w', edgecolor='w', orientation='portrait', papertype='a4', format='pdf', transparent=False, bbox_inches=None, pad_inches=0.1, frameon=None, metadata=None)
```

```python
time_data_outputs = actuator_outputs.data['timestamp']
# motors 2 & 3
left_motors = (actuator_outputs.data['output[1]'] + actuator_outputs.data['output[2]']) / 2
# motors 1 & 4
right_motors = (actuator_outputs.data['output[0]'] + actuator_outputs.data['output[3]']) / 2
fig, axs = plt.subplots(4, 1, sharex='all')
axs[0].plot(time_data, roll)
axs[3].plot(time_data_outputs, left_motors - right_motors)
axs[0].set_ylabel('Roll')
axs[0].grid(True)
axs[1].plot(time_data_outputs, left_motors)
axs[1].set_ylabel('Left motors')
axs[1].grid(True)
axs[2].plot(time_data_outputs, right_motors)
axs[2].set_ylabel('Right motors')
axs[2].grid(True)
fig.tight_layout()
# fig.savefig(f"axe_lateral.pdf", dpi=None, facecolor='w', edgecolor='w', orientation='portrait', papertype='a4', format='pdf', transparent=False, bbox_inches=None, pad_inches=0.1, frameon=None, metadata=None)
```

```python
thrust = (actuator_outputs.data['output[1]'] + actuator_outputs.data['output[2]'] + actuator_outputs.data['output[0]'] + actuator_outputs.data['output[3]']) / 4
plt.plot(time_data_outputs, thrust)
```
